```python
## 使用Sklearn调用GBDT模型拟合数据并可视化
import numpy as np
import pydotplus
from sklearn.ensemble import GradientBoostingRegressor

X = np.arange(1, 11).reshape(-1, 1)
print(X)
y = np.array([5.16, 4.73, 5.95, 6.42, 6.88, 7.15, 8.95, 8.71, 9.50, 9.15])
print(y)
gbdt = GradientBoostingRegressor(max_depth=3, criterion ='squared_error', n_estimators=3).fit(X, y)
```

```python
from IPython.display import Image  
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz

def plot_tree(i):
    sub_tree = gbdt.estimators_[i, 0]
    dot_data = export_graphviz(sub_tree, out_file=None, filled=True, rounded=True, special_characters=True, precision=2)
    return graph_from_dot_data(dot_data)  
```

```python
# mean + learning_rate * leaf_value
gbdt.predict(X)
```

```python
Image(plot_tree(0).create_png())
```
