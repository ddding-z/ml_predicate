```python
## 使用Sklearn调用GBDT模型拟合数据并可视化
import numpy as np
import pydotplus
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

X = np.arange(1, 11).reshape(-1, 1)
print(X)
y = np.array([0, 1, 2, 1, 2, 0, 2, 0, 1, 0])
print(y)
rf = RandomForestClassifier(max_depth=3, n_estimators=3).fit(X, y)
```

```python
from IPython.display import Image  
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz

def plot_tree(i):
    sub_tree = rf.estimators_[i]
    dot_data = export_graphviz(sub_tree, out_file=None, filled=True, rounded=True, special_characters=True, precision=2)
    return pydotplus.graph_from_dot_data(dot_data)  
```

```python
rf.predict(X)
```
