```python
file = 'branch_test.csv'
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv(file)
df
```

```python
left_data = df[df['left'] == True]['cost'].groupby(df['depth'], sort=True).mean().sort_index()
right_data = df[df['left'] == False]['cost'].groupby(df['depth'], sort=True).mean().sort_index()
right_data
```

```python
x = [i for i in range(1, 101)]
y1 = left_data.values
y2 = right_data.values
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y1, marker='o', label='left')
plt.plot(x, y2, marker='o', label='right')
plt.ylabel('time (s)')
plt.xlabel('depth')
plt.legend()
plt.ylim(0)
```

```python
plt.plot(x, y1 / y2, marker='o', label='left / right')
plt.legend()
plt.ylim(0)
```
